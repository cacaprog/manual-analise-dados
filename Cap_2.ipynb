### **Estatísticas Descritivas**

Estatísticas descritivas são usadas para descrever e resumir os dados. Elas fornecem uma visão geral dos principais aspectos de um conjunto de dados.
Variáveis Qualitativas

- Variáveis qualitativas (ou categóricas) são aquelas que descrevem qualidades ou categorias. Exemplos incluem gênero, cor dos olhos, e tipo de produto.

Principais Medidas para Variáveis Qualitativas

**Frequência Absoluta e Relativa:**
- Frequência Absoluta: O número de vezes que uma categoria ocorre no conjunto de dados.
- Frequência Relativa: A proporção de observações que uma categoria representa, calculada como a frequência absoluta dividida pelo total de observações.

**Moda:**
- A categoria mais frequente no conjunto de dados.

Variáveis Quantitativas

Variáveis quantitativas (ou numéricas) são aquelas que podem ser medidas numericamente. Exemplos incluem altura, peso e renda.
Principais Medidas para Variáveis Quantitativas

**Medidas de Tendência Central:**
- Média: A soma de todos os valores dividida pelo número de valores.
- Mediana: O valor do meio quando os dados são ordenados.
- Moda: O valor mais frequente (também aplicável a variáveis qualitativas).

**Medidas de Dispersão:**
- Desvio Padrão (Standard Deviation): Mede o quanto os valores variam em relação à média.
- Variância: O quadrado do desvio padrão.
- Amplitude (Range): A diferença entre o maior e o menor valor.
- Erro Padrão (Standard Error): ele estima a variabilidade das médias amostrais se repetíssemos a amostragem várias vezes. Quanto menor o erro padrão, mais precisa é a estimativa da média da população.
- Desvio do Erro Padrão (Standard Error of the Mean Deviation): O desvio do erro padrão geralmente se refere ao erro padrão de uma medida específica, como a diferença entre médias ou outras estatísticas derivadas. Quando se trabalha com comparações entre grupos, o erro padrão das diferenças também é importante.

Interpretação dos Resultados
- Frequência: Ajuda a entender a distribuição de categorias. Frequências relativas são úteis para comparações.
- Tendência Central: Média, mediana e moda oferecem diferentes perspectivas sobre o "centro" dos dados. A média pode ser afetada por valores extremos, enquanto a mediana não.
- Dispersão: Desvio padrão e variância mostram a variação dos dados em relação à média. Amplitude indica a faixa de valores.
- Erro Padrão: Um menor erro padrão indica que a média amostral está mais próxima da verdadeira média populacional. Usado em intervalos de confiança e testes de hipóteses para determinar a significância estatística.

**IQR**

O IQR é particularmente útil em situações onde queremos entender a variabilidade dos dados sem ser influenciado por outliers extremos. Se precisar de mais detalhes ou ajuda com outro ponto específico, estou à disposição para ajudar

- IQR: É uma medida da dispersão central dos dados. Um IQR maior indica que os dados estão mais dispersos em torno da mediana.
- Outliers: Qualquer valor abaixo de Q1 − 1.5 × IQR ou acima de Q3 + 1.5 × IQR é considerado um outlier


**Coeficiente de Assimetria de Fisher (g1)**

O coeficiente de assimetria de Fisher (g1) mede a simetria da distribuição dos dados. 
- Um valor de g1 perto de zero indica uma distribuição simétrica. 
- Valores positivos indicam assimetria à direita (distribuição inclinada para a direita), 
- Valores negativos indicam assimetria à esquerda (distribuição inclinada para a esquerda).


**Coeficiente de Curtose de Fisher (g2)**

O coeficiente de curtose de Fisher (g2) mede a "pontualidade" da distribuição dos dados. 
- Uma curtose de 3 (ou g2 igual a 0) indica uma distribuição normal (mesocúrtica). 
- Valores positivos indicam uma distribuição mais pontuda (leptocúrtica)
- Valores negativos indicam uma distribuição mais achatada (platicúrtica).

```python
# questão 12-h
# o histograma já foi construído anteriormente,
# vamos então criar o gráfico de ramo-e-folhas e o boxplot para a variável em estudo

# Função para criar o gráfico de ramos e folhas
def stem_and_leaf(data):
    stems = {}
    for number in sorted(data):
        stem = number // 5
        leaf = number % 20
        if stem in stems:
            stems[stem].append(leaf)
        else:
            stems[stem] = [leaf]
    
    # Criar a representação textual do gráfico
    for stem, leaves in stems.items():
        print(f"{stem} | {' '.join(map(str, leaves))}")

# Criar o gráfico de ramos e folhas
stem_and_leaf(desidratacao["segundos"].round(1))
```

```python
"""
Significado de ddof
    ddof significa "delta degrees of freedom" (delta graus de liberdade).
    O divisor utilizado no cálculo é N - ddof, onde N é o número de elementos na amostra.

Por que ddof=1?
Quando usamos ddof=1, estamos aplicando a correção de Bessel. Isso ajusta a fórmula para 
calcular a variância e o desvio padrão de uma amostra, tornando-a uma estimativa imparcial 
da variância e do desvio padrão da população.

Interpretação
Sem correção de Bessel (ddof=0): O desvio padrão é calculado usando o divisor N.
Com correção de Bessel (ddof=1): O desvio padrão é calculado usando o divisor N-1, 
o que geralmente é mais apropriado para amostras.

Essa correção é especialmente importante quando se trabalha com amostras pequenas, 
pois ajuda a evitar a subestimação da variância e do desvio padrão da população.
"""

servicos_desvio_padrao_2 = np.std(servicos, ddof=1)
print(servicos_desvio_padrao_2, servicos_desvio_padrao)
```

Coeficiente de Assimetria de Fisher (g1) mede a simetria da distribuição dos dados.

- g1 = 0 : distribuição simétrica
- g1 > 0 : assimetria à direita (distribuição inclinada à direita)
- g1 < 0 : assimetria à esquerda (distribuição inclinada à esquerda)

O coeficiente de curtose de Fisher (g2) mede a "pontualidade" da distribuição dos dados. 

- g2 = 0 : indica uma distribuição normal (mesocúrtica). 
- g2 > 0 : distribuição mais pontuda (leptocúrtica)
- g2 < 0 : indicam uma distribuição mais achatada (platicúrtica)
